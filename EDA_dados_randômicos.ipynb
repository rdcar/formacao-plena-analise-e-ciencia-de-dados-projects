```python
#Importando biblioteca
import numpy as np
```

## Gerando listas com alturas, pesos, IMCs e Sexo

```python
#Gerando dados de média 1.75 e desvio padrão 0.30, sendo 2000 pessoas em uma distribuição normalizada.
alturas = np.random.normal(loc=1.75, scale=0.3, size=2000)
alturas = alturas.tolist() #Arrays não permitem manipulação, por isso, precisamos transformar em list
#Usando os valores da lista Alturas, multiplicando por 50 e adicionando um número aleatório entre -5 e 5 para criar o dado de peso, apendado a lista Pesos
pesos = []
for i in alturas:
  a = i*50
  b = float(np.random.uniform(-5,5))
  c = a + b
  pesos.append(c)
# Criando lista com imc
imc = []
for i, j in zip(pesos,alturas):
  x = i/(j**2)
  x = round(x,2)
  imc.append(x)
# Criando lista com gênero
genero = []
for i in range(0,2000):
  x=np.random.uniform(0,1)
  if x < 0.5:
    genero.append("M")
  else:
    genero.append("F")
# Criando lista com pessoas cardiopatas (S), possivelmente cardiopatas(P) ou sem problemas cardíacos (N).
cardiaco = []
for i in range(0,2000):
  if imc[i] <= 30:
    cardiaco.append("N")
  elif imc[i] < 40:
    cardiaco.append("P")
  elif imc[i] >=40:
    cardiaco.append("S")
```

```python
# Salvando todas as listas em uma lista única (lista de listas)
all_lists = alturas, pesos, imc, genero, cardiaco
```

```python
import pandas as pd
# Transformando a lista de listas criada em um DataFrame
df = pd.DataFrame(all_lists, index=["Altura","Peso","IMC","Gênero","Cardíaco"])
df.head()
```

```python
# Transpondo linhas x colunas
df = pd.DataFrame(all_lists, index=["Altura","Peso","IMC","Gênero","Cardíaco"]).T
df
```
